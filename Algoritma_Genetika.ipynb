```python
import random

# Fungsi yang ingin dimaksimalkan
def fitness(x):
    return x**2 + 2

# Inisialisasi populasi
def init_population(pop_size, lower_bound, upper_bound):
    return [random.uniform(lower_bound, upper_bound) for _ in range(pop_size)]

# Seleksi berdasarkan fitness
def selection(population):
    population.sort(key=lambda x: fitness(x), reverse=True)
    return population[:3]  # Memilih 3 terbaik

# Crossover sederhana dengan rata-rata
def crossover(x1, x2):
    return (x1 + x2) / 2

# Mutasi dengan menambahkan nilai kecil
def mutation(x, mutation_rate):
    if random.random() < mutation_rate:
        return x + random.uniform(-0.5, 0.5)  # Mutasi kecil
    return x

# Parameter
pop_size = 5
num_generations = 100
crossover_rate = 0.5
mutation_rate = 0.1
lower_bound = -4
upper_bound = 4

# Inisialisasi populasi
population = init_population(pop_size, lower_bound, upper_bound)

# Algoritma Genetika
for generation in range(num_generations):
    # Evaluasi fitness dan seleksi
    selected = selection(population)

    # Crossover
    if random.random() < crossover_rate:
        x1, x2 = random.sample(selected, 2)
        offspring = crossover(x1, x2)
        selected.append(offspring)

    # Mutasi
    selected = [mutation(x, mutation_rate) for x in selected]

    # Populasi baru
    population = selected + init_population(pop_size - len(selected), lower_bound, upper_bound)

# Output hasil terbaik
best_individual = max(population, key=fitness)
print(f"Nilai maksimum x: {best_individual}, y: {fitness(best_individual)}")
```

```python
import random

# Fungsi yang ingin dimaksimalkan
def fitness(x):
    return x**2 + 2

# Inisialisasi populasi
def init_population(pop_size, lower_bound, upper_bound):
    return [random.uniform(lower_bound, upper_bound) for _ in range(pop_size)]

# Seleksi berdasarkan fitness
def selection(population):
    # Mengurutkan populasi berdasarkan nilai fitness (descending)
    population.sort(key=lambda x: fitness(x), reverse=True)
    return population[:3]  # Memilih 3 terbaik

# Crossover sederhana dengan rata-rata
def crossover(x1, x2):
    return (x1 + x2) / 2

# Mutasi dengan menambahkan nilai kecil
def mutation(x, mutation_rate):
    if random.random() < mutation_rate:
        return x + random.uniform(-0.5, 0.5)  # Mutasi kecil
    return x

# Parameter
pop_size = 5
num_generations = 100
crossover_rate = 0.5
mutation_rate = 0.1
lower_bound = -4
upper_bound = 4

# Inisialisasi populasi
population = init_population(pop_size, lower_bound, upper_bound)

# Algoritma Genetika
for generation in range(num_generations):
    print(f"Generasi {generation + 1}")

    # Evaluasi fitness dan seleksi
    selected = selection(population)
    print(f"Individu terpilih: {selected}")

    # Crossover
    if random.random() < crossover_rate:
        x1, x2 = random.sample(selected, 2)
        offspring = crossover(x1, x2)
        selected.append(offspring)
        print(f"Offspring dari crossover: {offspring}")

    # Mutasi
    selected = [mutation(x, mutation_rate) for x in selected]
    print(f"Individu setelah mutasi: {selected}")

    # Populasi baru
    population = selected + init_population(pop_size - len(selected), lower_bound, upper_bound)
    print(f"Populasi baru: {population}\n")

# Output hasil terbaik setelah semua generasi
best_individual = max(population, key=fitness)
print(f"Nilai maksimum x: {best_individual}, y: {fitness(best_individual)}")
```
